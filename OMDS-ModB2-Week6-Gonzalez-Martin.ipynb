# 2. Weekly graph question

Suppose you wanted to show the nitrate and phosphate level in a water sample.  What are the advantages and disadvantages of showing this as an area plot, as opposed to two separate line graphs?  How would you adjust the graphs shown to improve the presentation?

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series_2 =  100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series_2)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

```python
df = pd.DataFrame({'ser1': time_series, 'ser2': time_series_2}, index = datearray)
df.plot.area()
```

```python

df = pd.read_csv("diabetes_binary_health_indicators_BRFSS2015.csv", sep=",")

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in df.columns:
    plt.figure(figsize=(6,4))
    sns.histplot(data=df, x=col, bins=20, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

The histrograms show the distrbutions in the data. We see from education and income there are more individuals with a higher income (grade 8: $75,000+) and from edcation there are individuals with a higher form of education.

```python
# find the correlation

df.corr()
```

```python
# heatmap

plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation using a Heatmap")
plt.show()
```

This heat map gives us an easier view of what columns are highly correlated. From this we can see more variables are negatively correlated rahter than positively correlated

```python
corr_matrix = df.corr().abs()

np.fill_diagonal(corr_matrix.values, 0) # set the diagnoial to 0 so it's ignored
corr_pairs = corr_matrix.unstack()
sorted_pairs = corr_pairs.sort_values(ascending=False)

print(sorted_pairs.head(10))
```

```python
corr_matrix = df.corr().abs()

np.fill_diagonal(corr_matrix.values, 0) # set the diagnoial to 0 so it's ignored
corr_pairs = corr_matrix.unstack()
sorted_pairs = corr_pairs.sort_values(ascending=False)

print(sorted_pairs.head(10))
```

```python
top_corr_vars = ["PhysHlth", "GenHlth", "DiffWalk", "Education", "Income"]

sns.pairplot(df[top_corr_vars], diag_kind="kde", corner=True)
plt.title("Pairplot of correlated variables")
plt.show()
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 86-98 of the Storytelling With Data book as best you can.  (The second half of chapter three).  You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

ticket_received = [160, 175, 252, 160, 165, 150, 210, 160, 150, 155, 175]
ticket_processed = [160, 175, 250, 160, 150, 140, 160, 145, 130, 135, 150]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov"]

plt.figure(figsize=(10, 6))
plt.plot(months, ticket_received, label='Ticket Volume Received', color='blue')
plt.plot(months, ticket_processed, label='Ticket Volume Processed', color='red')
plt.title("Ticket Volume Received vs Processed (Jan–Nov)")
plt.xlabel("Month")
plt.ylabel("Ticket Volume")
plt.ylim(0, 300)
plt.xlim(0,11)
plt.grid(False)
plt.legend()
plt.tight_layout()
plt.show()
```
